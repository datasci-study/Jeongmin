```python
anyang_gid=anyang_pop.drop(['val'], axis = 1)

def df_to_geojson(df, properties, lat = 'lat', lon = "lon"):
    geojson = {"type" : 'FeatureCollection',"name":'floating', 
               "crs": { "type": "name", "properties": { "name": "urn:ogc:def:crs:OGC:1.3:CRS84" } }
               ,"features":[]}
    for _, row in df.iterrows():
        feature = {'type' : 'Feature',
                  'properties':{},
                  'geometry':{"type":"Point", "coordinates":[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop]=row[prop]
        geojson['features'].append(feature)
    return geojson

def writinggeojson(df, properties, filename):
    save_file = filename
    setjson = df_to_geojson(df, properties)
    
    with open(save_file,'w',encoding='utf-8') as f:
        dump(setjson, f)

def labeledgeojson(df1, anyang_gid, col):
    re_data = gpd.sjoin(df1, anyang_gid, how = 'right', op = 'within')
    gid_data = re_data.groupby('gid').sum()
    gid_data.drop(['index_left'], axis = 1)
    gid_data = gid_data.reset_index().copy()
    gid_data = gid_data[[col, "gid"]]
    geo_gid_data = gid_data.merge(anyang_gid, on ='gid')
    
    anyang_copy = geo_gid_data.copy()
    breaks = jenkspy.jenks_breaks(anyang_copy[col], nb_class = 5)
    anyang_copy['class'] = pd.cut(anyang_copy[col],
                                 bins = breaks,
                                 labels = [1,2,3,4,5],
                                 include_lowest = True)
    return anyang_copy
```
